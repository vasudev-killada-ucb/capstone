```
import boto3
import email
from email import policy
from email.parser import BytesParser
import uuid

s3 = boto3.resource('s3')
s3_client = boto3.client('s3')
comprehend = boto3.client('comprehend', region_name='us-west-2')

bucket_name = 'capstone-enron'
folder_name = 'campbell-l/_sent_mail/'

bucket = s3.Bucket(bucket_name)
emails = []

for obj in bucket.objects.filter(Prefix=folder_name):
    key = obj.key
    email_object = s3_client.get_object(Bucket=bucket_name, Key=key)
    email_bytes = email_object['Body'].read()
    message = BytesParser(policy=policy.default).parsebytes(email_bytes)
    sender_id = str(uuid.uuid4())
    sender_email = message['From']
    email_content = message.get_body().get_content()
    
    if len(email_content) > 5000:
        email_chunks = [email_content[i:i+5000] for i in range(0, len(email_content), 5000)]
        for i, chunk in enumerate(email_chunks):
            chunk_id = f"{sender_id}_{i}"
            emails.append({
                'sender_id': chunk_id,
                'sender_email': sender_email,
                'email_content': chunk
            })
    else:
        emails.append({
            'sender_id': sender_id,
            'sender_email': sender_email,
            'email_content': email_content
        })

batch_size = 25
email_batches = [emails[i:i + batch_size] for i in range(0, len(emails), batch_size)]
for batch in email_batches:
    batch_content = [email['email_content'] for email in batch]
    sentiments = comprehend.batch_detect_sentiment(
        TextList=batch_content,
        LanguageCode='en'
    )
    for i, email in enumerate(batch):
        if '_' in email['sender_id']:
            sender_id, chunk_num = email['sender_id'].split('_')
            email_sentiment = sentiments['ResultList'][i]['Sentiment']
            sentiment_score = sentiments['ResultList'][i]['SentimentScore']
            print(f"Sender: {email['sender_email']}, ID: {sender_id}, Chunk: {chunk_num}, email_body: {email_content}, Sentiment: {email_sentiment}, SentimentScore: {sentiment_score}")
        else:
            email_sentiment = sentiments['ResultList'][i]['Sentiment']
            sentiment_score = sentiments['ResultList'][i]['SentimentScore']
            print(f"Sender: {email['sender_email']}, ID: {email['sender_id']}, email_body: {email_content},Sentiment: {email_sentiment}, , SentimentScore: {sentiment_score}")
```
